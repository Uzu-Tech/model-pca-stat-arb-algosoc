```python
from pipline import data_loader
from alpaca.data.timeframe import TimeFrame
import json
```

```python
BACK_TEST_YEARS = 2
TIMEFRAME: TimeFrame = TimeFrame.Hour # type: ignore

tickets_filename = "edufund_tickets.json"

# Excluding any market or sector stocks since we just want individual stocks for PCA
etfs_to_exclude = {
    "SPY", "QQQ", "IWM", "DIA", "VOO", "VTI", "XLE", "XLF", "XLK", 
    "XLP", "XLU", "XLV", "XLY", "XLI", "XLB", "XLRE", "EFA", "EEM", 
    "VEA", "AGG", "LQD", "HYG", "TLT", "IEF", "BND", "EMB", "GLD", 
    "SLV", "VNQ"
}

with open(data_loader.DATA_PATH / tickets_filename, mode="r") as f:
    all_tickets = json.load(f)

tickets = [t for t in all_tickets["equity_symbols"] if t not in etfs_to_exclude]

df = data_loader.load_stock_data(
    tickets, years_from_2025_end=BACK_TEST_YEARS, timeframe=TIMEFRAME
)

df
```
